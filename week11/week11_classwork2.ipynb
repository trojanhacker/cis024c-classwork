```python
# line chart
import matplotlib.pyplot as plt

x = [1,2,3,4,5]

y1 = [1,4,9,16,25]
y2 = [1,8,27,64,125]

plt.plot(x,y1,label='first line',color='red')
plt.plot(x,y2,label='second line',color='black')
plt.xlim(0,3)

plt.title('my first plot')
plt.xlabel('indices')
plt.ylabel('response values')

plt.legend()

plt.show()
```

```python
# Subplots

import matplotlib.pyplot as plt
import numpy as np



x = np.linspace(1,100,100)
y = np.random.normal(179,29,100)

f,asubplot = plt.subplots(2,sharex=True,figsize=(10,10))



asubplot[0].plot(x,y,color='r')
asubplot[0].set_title('Line Chart')

asubplot[1].hist(y,color='b')
asubplot[1].set_title('Histogram Chart')

plt.show()

```

```python
import matplotlib.pyplot as plt
import csv

admit = []
gender = []
dept = []
freq = []

with open('UCBAdmissions.csv','r') as csvFile:
    
    admission_stats = csv.reader(csvFile,delimiter=',')
    
    firstTime = True
    for row in admission_stats:
        
        if firstTime == True:
            firstTime = False
            continue
            
        admit.append(row[1].strip("'"))
        gender.append(row[2].strip("'"))
        dept.append(row[3].strip("'"))
        freq.append(int(row[4]))
        
print admit,gender,dept,freq
        
plt.hist(freq)
```

```python
import matplotlib.pyplot as plt

aliceFile = open('aliceText.txt')

wordCounts = {}

for line in aliceFile:
    
    words = line.strip().split(' ')
    for word in words:
        
        wordLength = len(word)
        
        if wordCounts.get(wordLength) == None:
            wordCounts[wordLength] = 1
        else:
            wordCounts[wordLength] += 1
            
wordLengths = []

for wordlen in wordCounts:
    wordLengths += [wordlen]*wordCounts[wordlen]

plt.xlim(0,20)
plt.hist(wordLengths)
    
    
        
```
